# Tablas de contingencia para 2 variables

Frecuencias absolutas

```python
survived_sex = pd.crosstab(index = titanic['survived'],
                             columns = titanic['sex'])
survived_sex.index = ["died", "survived"] # etiquetas nuevas para las filas
survived_sex
```

```python
survived_class = pd.crosstab(index = titanic['survived'], 
                             columns = titanic ['pclass'],
                             margins = True)
survived_class.index = ["murio", "sobrevivió", "total_clase"]
survived_class.columns = ["primera", "segunda", "tercera","total_superv"]
survived_class
```

## Frecuencias relativas globales

```python
survived_class/survived_class.loc["total_clase", "total_superv"] # se pone loc para localizar [filas, columnas]
```

## Frecuencias relativas marginales

```python
survived_class/survived_class.loc["total_clase"] # por filas
```

```python
survived_class.div(survived_class.loc["total_clase"], axis=1 ) # otra forma, por filas
```

```python
survived_class.T/survived_class["total_superv"] # por columnas, con la Traspuesta
```

```python
survived_class.div(survived_class["total_superv"],axis = 0) # otra forma,con .div por columnas
```

## Tablas multidimensionales

```python
surv_sex_class = pd.crosstab(index = titanic ["survived"],
                             columns = [titanic["sex"], titanic["pclass"]],
                             margins = True)
surv_sex_class # hemos creado una tabla multidimensional con 3 factores:1 fila y 2 columnas
                            
```

```python
surv_sex_class["female"] # para ver las mujeres
```

```python
surv_sex_class["female"][1] # mujeres en 1ª clase, debe ir en el mismo orden que en la tabla de estudio
```
