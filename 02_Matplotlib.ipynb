# Múltiples gráficos en una misma figura

```python
import numpy as np
```

```python
def f(x):
  return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5.0, 0.1)
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x1, f(x1), "ro", x2, f(x2), "k")

plt.subplot(212)
plt.plot(x2, f(x2),"g--")

plt.show()
```

```python
plt.figure(1) # creamos una figura con dos subplots
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2) #una sola figura con 1 subplot, sale más grande
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title("Esto es el primer título")#para poner título
```

# Textos en gráficos

```python
# Vamos a crear una distribución Normal aleatoria
mu = 100       # media
sigma = 20     # desviación típica
x = mu + sigma *np.random.randn(10000)
```

```python
from IPython.core.display import Latex
# density para frecuencias relativas
# sin density para frecuencias absolutas
n, bins, patches = plt.hist(x, 50, density = 1, facecolor = "teal", alpha = 0.6) 

plt.xlabel("Cociente intelectual", fontsize = 14, color = "b")
plt.ylabel("Probabilidad", fontsize = 14, color = "b")
plt.title(r"Histograma de CI  $N(\mu, \sigma)$", fontsize = 20, color = "r")
plt.text(30, 0.015, r'$\mu = 100, \ \ sigma = 20$') # r indica que es formato Latex
plt.axis([20,180,0,0.025]) #ajuste de los ejes
plt.grid(True) #lineas divisorias

plt.show()
```

```python
# para añadir objetos como flechas, con annotate()
# figsize tamaño figura
# nº pixeles por pulgada, definición

plt.figure(figsize = (10,6), dpi = 90) 
plt.subplot(1,1,1)

x = np.arange(0,10*np.pi, 0.01)
y = np.cos(x)

plt.plot(x,y, lw = 2.0)
plt.annotate('Máximo local', xy =(4*np.pi,1), xytext = (15, 1.5), arrowprops = dict(facecolor = "black", shrink = 0.05))

plt.ylim(-2,2)

plt.show()
```

# Cambios de ejes y escalas

```python
from matplotlib.ticker import NullFormatter # para poder modificar la visualización de gráficos
```

```python
# vamos a crear 1000 números aleatorios
mu = 0.5
sd = 0.3
y = mu + sd * np.random.randn(1000)
y = y [(y > 0) & (y < 1)] #acotamos, es un vector unidimensional, no hace falta poner comas como hacíamos en R
y.sort()
x = np.arange(len(y))
```

```python
# vamos a crear los 4 ejes en la misma figura

plt.figure(figsize = (10, 8))

plt.subplot(221)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")
plt.grid(True)

plt.subplot(222)
plt.plot(x,y)
plt.yscale("log") #semilogarítmica vamos a hacer en este caso
plt.title("Escala Logarítmica")
plt.grid(True)

plt.subplot(223)
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthreshy = 0.01)
plt.title("Escala Logaritmo Simétrico")
plt.grid(True)

plt.subplot(224)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) # no sirve
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)

plt.show()
```

```python
plt.plot(x, S, color = "blue", lw = 1.2, linestyle = "-", label = "seno") #es lo mismo que (x, 1ª f(x))
plt.plot(x, C, color = "orange", linewidth = 1.2, linestyle = "-", label = "coseno") #(x, 2ªf(x))
plt.xlim(-4,4) #plt.xlim(x.min()*1.1, x.max()*1.1) para gestionar dinámicamente
plt.ylim(-1.2,1.2) #plt.ylim(S.min()*1.1, S.max()*1.1) 
plt.xticks(np.linspace(-4,4,9, endpoint = True))
plt.yticks(np.linspace(-1,1,9, endpoint = True))
plt.show()
```

```python
plt.plot(x, S, color = "blue", lw = 1.2, linestyle = "-", label = "seno") #es lo mismo que (x, 1ª f(x))
plt.plot(x, C, color = "orange", linewidth = 1.2, linestyle = "-", label = "coseno") #(x, 2ªf(x))
plt.xlim(x.min()*1.1, x.max()*1.1) #para gestionar dinámicamente
plt.ylim(S.min()*1.1, S.max()*1.1) 
#si queremos poner los ejes con múltiplos de pi:
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])
plt.yticks(np.linspace(-1,1, 2, endpoint=True),
          ['-1',  '+1'])
# para borrar los bordes del marco usamos ax.spines y color none:
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# para situar el eje de coordenadas (0,0):
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.legend(loc="upper left")
 
# para hacer marcas dentro del gráfico
# una línea que marque un pto concreto en azul
x0 = 2*np.pi/3
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--")
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue")
# anotación con la ecuación en unas coordenadas concretas y una flecha
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20,+40), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2"))
# una línea que marque un pto concreto en verde
plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--")
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green")
# anotación con la ecuación del coseno
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2"))
# etiquetas con un fondo blanco y mayor tamaño:
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='whitesmoke', edgecolor='None', alpha = 0.6))
    # probar con label.set_bbox(dict(facecolor='red', edgecolor='None', alpha = 0.6))
    # así ver bien que hacen las etiquetas


plt.show()
```
