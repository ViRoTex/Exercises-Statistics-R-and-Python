# Pregunta 3
Calcula el porcentaje medio de variación del pulso por minuto entre antes y después de hacer ejercicio y
compara el valor de los que hacen ejercicio habitualmente y los que no. ¿Observas mucha diferencia?

```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/run.csv')
df
```

```python
# cambiamos los ptos de las columnas por barras bajas 
df.columns = ['indice','pulso_antes','pulso_despues','hace_deporte','fuma','genero' ,'altura','peso', 'tipo_actividad']
df
```

```python
# creamos una función que nos dé el %pulso medio
def porcent_var_pulso(x):
  x['%pulso_medio'] = (x['pulso_despues']- x['pulso_antes'])/ x['pulso_antes']
  return x
```

```python
# aplicamos dicha función a la tabla creada para añadir una nueva columna con el resultado
df = df.apply(porcent_var_pulso, axis = 1).round(2)
```

```python
# hay que agrupar según si hacen o no deporte

df = df.set_index('hace_deporte')
df
```

```python
#vamos a hacer otra función con la media de los que hace deporte con los que no

df.groupby('hace_deporte').aggregate([np.mean]).round(2)
```
