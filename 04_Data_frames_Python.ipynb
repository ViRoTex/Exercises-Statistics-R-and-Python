# Manipulación o agregados de datos en DF

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'value1' : range(6),
    'value2' : np.random.randint(0,100,6)
})
```

```python
df
```

## Aggregate
Aggregate se puede poner como:

*   un vector con string o funciones: ['min',np.mean, max]
*   o como un diccionario: {filas:columnas}





```python
df.groupby('key').aggregate(['min', np.median, np.mean, max]) 
```

```python
df.groupby('key').aggregate([min, np.median, np.mean, max]) #la función min y max da igual si se pone como string o no
```

## Filter

```python
df.groupby('key').std()
```

```python
def filter_func(x):
  return x['value2'].std() > 25
```

```python
#from IPython.display import display

#display('df', 'df.groupby('key').std()', 'df.groupby('key').filter(filter_func)')
```

```python
df.groupby('key').filter(filter_func) # nos elimina la columna C por no cumplir la condición booleana
```

## Transform
cuando queremos la versión reducida de los datos pero sin combinar, es decir, sin cambiar el tamaño original(filter modificaba o reducía el tamaño de los datos de la tabla)

```python
df.groupby('key').transform(lambda x : x - x.mean())
```

## Apply
permite pasar una función y aplicarla en cada valor (nos ahorra hacer un for....in....:)

```python
def norm_by_col2(x):
  x['value1'] /= x['value2'].sum() # es lo mismo que x['value1'] = x['value1']/x['value2'].sum()
  return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

# División de datos por columnas. Parámetros de la función **groupby**

```python
L = [0,1,0,1,2,0]
print(df)
```

```python
df.groupby(L).sum() #agrupación elegida por mí con L
```

```python
df.groupby('key').sum() # agrupación según la columna 'key'
```

```python
df.groupby(df['key']).sum() #agrupación según vector, en este caso la columna 'key como vector
```

```python
df2 = df.set_index('key') #quitamos el índice y ponemos key como índice
df2
```

```python
mapping = {      #diccionario creado para agrupar las filas 
    'A':'vowel',
    'B':'consonant',
    'C':'consonant'
}
```

```python
df2.groupby(mapping).sum() #agrupación por diccionario
```

```python
df2.groupby(str.lower).mean() #con funciones lower y mean
```

```python
df2.groupby([str.lower, mapping]).mean()
```

# Gráficos desde Data Frames y Exportación

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
years = [year for year in range(1900, 2020)] #creamos 2 arrays
deads = [(y + np.random.uniform(0,100)-1850) for y in years ]
```

```python
df = pd.DataFrame({ #los unimos en un DF
    "year":years,
    "deads":deads
})
```

```python
df.head()
```

```python
df.plot(x="year", y ="deads", figsize = (10,8)) #hacemos un plot
```

```python
df.plot(kind = "scatter", x="year", y ="deads", figsize = (10,8)) # si lo queremos en ptos dispersos
```

```python
#vamos a guardar el gráfico de líneas en nuestro pc
my_plot = df.plot(x="year", y ="deads", figsize = (10,8)) # le asignamos el nombre my_plot
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("/content/data_frame_plot.png")
```
